> ***Analisando idades mínimas, média e máxima de clientes respectivamente:***

**Query:** SELECT MIN(idade) AS minima_idade, AVG(idade) AS media_idade, MAX(idade) AS maxima_idade FROM credito;



> ***Agrupando as informações anteriores por sexo:***

**Query:** SELECT MIN(idade) AS minima_idade, AVG(idade) AS media_idade, MAX(idade) AS maxima_idade, sexo FROM credito GROUP BY sexo;



* Esses dados não forneceram nenhuma informação relevante.


> ***Analisando mínima e máxima transações de clientes em 12 meses:***

**Query:** SELECT MIN(valor_transacoes_12m) AS minima_transacoes, MAX(valor_transacoes_12m) AS maxima_transacoes FROM credito;



* As transações variam entre 510.16 e 4776.58.



> ***Analisando limite de crédito de clientes baseado em sua escolaridade:***

**Query:** SELECT MAX(limite_credito) AS limite_credito, escolaridade
FROM credito 
WHERE escolaridade != 'na' 
GROUP BY escolaridade 
ORDER BY limite_credito;



* Esses dados não forneceram nenhuma informação relevante.


> ***Analisando limite de crédito entre homens e mulheres:***

**Query:** SELECT MAX(limite_credito) AS limite_credito, sexo FROM credito GROUP BY sexo ORDER BY limite_credito;



* O sexo dos clientes também não oferece nenhuma informação relevante.


> ***Analisando limite de crédito por tipo de cartão do cliente:***

**Query:** SELECT MAX(limite_credito) AS limite_credito, tipo_cartao FROM credito GROUP BY tipo_cartao ORDER BY limite_credito;



* Esses dados não forneceram nenhuma informação relevante.



> ***Analisando limite de crédito de clientes baseado em seu salário anual:***

**Query:** SELECT MAX(limite_credito) AS limite_credito, salario_anual FROM credito WHERE salario_anual != 'na' GROUP BY salario_anual  ORDER BY limite_credito;



* O limite de crédito são semelhantes entre clientes que ganham entre 60k a 120k anualmente.
* Os clientes que possuem salário anual inferior a 40k possuem um limite bem abaixo dos demais clientes
* **Nessa análise os clientes que não passaram informação, ou seja, que tinham os valores equivalente a "na" não estão sendo computados.**


> ***Analisando mínimas e máximas transações de clientes em 12 meses agrupando-os por salário anual:***

**Query:** SELECT MIN(valor_transacoes_12m) AS minimo_valor, MAX(valor_transacoes_12m) AS maximo_valor, salario_anual FROM credito WHERE salario_anual != 'na' GROUP BY salario_anual;
