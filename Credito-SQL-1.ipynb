# 1. Identificação de colunas


> ***Os dados representam informações de clientes de um banco e contam com as seguintes colunas:***

idade = idade do cliente

sexo = sexo do cliente (F ou M)

dependentes = número de dependentes do cliente

escolaridade = nível de escolaridade do clientes

salario_anual = faixa salarial do cliente

tipo_cartao = tipo de cartao do cliente

qtd_produtos = quantidade de produtos comprados nos últimos 12 meses

iteracoes_12m = quantidade de iterações/transacoes nos ultimos 12 meses

meses_inativo_12m = quantidade de meses que o cliente ficou inativo

limite_credito = limite de credito do cliente

valor_transacoes_12m = valor das transações dos ultimos 12 meses

qtd_transacoes_12m = quantidade de transacoes dos ultimos 12 meses

# 2. Exploração de dados:

> ***Verificando a quantidade de informações na base de dados:***

**Query:** SELECT count(*) FROM credito
Resposta: 2564 linhas


> ***Extraindo base de dados:***

**Query:** SELECT * FROM credito;




> ***Utilizando o método DESCRIBE para identificar o tipo de dados nas colunas:***

**Query:** DESCRIBE credito;



> ***Utilizando método DISTINCT pra entender melhor as colunas.***

Este método é utilizado para mostrar os valores, de forma única, de cada coluna selecionada e consequentemente identificando todos suas póssiveis variáveis;

> ***Identificando valores na coluna "escolaridade":***

**Query:** SELECT DISTINCT escolaridade FROM credito;



> ***Identificando valores na coluna "estado_civil":***

**Query:** SELECT DISTINCT estado_civil FROM credito;



> ***Identificando valores na coluna "salario_anual":***

**Query:** SELECT DISTINCT salario_anual FROM credito;









# 3. Análise de Dados


> ***Depois de analisar e tratar os dados é possível analisar informações no banco de dados.***
>
> Essas informações potencializam a tomada de decisão mais assertiva.
>
> ***Analisando a quantidade de clientes e o respectivo salário anual;***

**Query:** 
SELECT COUNT(*), salario_anual 
FROM credito 
GROUP BY salario_anual;



> ***Em alguns casos é possível realizar gráficos para melhor visualização e análise:***




Primeiras análises:
* 27% dos clientes possuem renda inferior a 40k anual.
* 9% dos clientes não passaram a informação sobre o salário anual.
* 9% dos clientes possuem uma renda superior a 120k anual.

> ***Analisando a quantidade de clientes agrupados por sexo;***

**Query:** SELECT COUNT(*), sexo FROM credito GROUP BY sexo;






* 61% dos clientes são do sexo masculino;


> ***Analisando o estado cívil dos clientes;***

**Query:** SELECT COUNT(*), estado_civil FROM credito GROUP BY estado_civil;





* Quase 60% dos clientes são casados.
* Apenas 5% dos clientes não passaram informação.


> ***Analisando escolaridade de clientes;***

**Query:** SELECT COUNT(*), escolaridade FROM credito GROUP BY escolaridade;





* 32% dos clientes tem mestrado e 28% doutorado
* 14% dos clientes não tem educação formal
* 14% dos clientes também não passaram a informação


> ***Analisando quantidade de clientes por tipo de cartão:***

**Query:** SELECT COUNT(*), tipo_cartao FROM credito GROUP BY tipo_cartao;





* 96% dos clientes utilizam o tipo de cartão "Blue".
* Existem apenas 2 clientes em todo banco de dados que utilizam o cartão "Platinum"
